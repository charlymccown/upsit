#### 2) Probability distributions for the statistical model's predicted probability of a given diagnostic label, separated post hoc into subjects that did not--by pathological assessment--have that label (yellow), and those that did (blue). ## Table 1

```python
Table('all_data',12)
```

#### T1) The statistical model uses dozens of features (out of over hundred available) to predict post-mortem pathologies, and for each diagnosis (column headers), the top 10 such features are indicated.  Features that were especially important are indicated with increased font size.  Smell-related features are indicated in red.  These smell-related features (largely UPSIT individual question responses) make up a large fraction of the top 10 most diagnostic features for many of the pathologies.
